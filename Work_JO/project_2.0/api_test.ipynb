```python
def get_json(count, district=None):
    if district == None:
        url = f'http://openapi.seoul.go.kr:8088/74684c474b776c6438354f50537661/json/IotVdata017/1/{count}/'
    else:    
        url = f'http://openapi.seoul.go.kr:8088/74684c474b776c6438354f50537661/json/IotVdata017/1/{count}/{district}'

    import requests
    row_data = requests.get(url).json()
    json_data = row_data['IotVdata017']['row']

    return json_data
```

```python
json_data = get_json(1000)

json_data
```
